```python
import numpy as np
import pandas as pd

n = 50
data = {'test1': np.random.normal(14, 3, n),
        'test2': np.random.normal(96, 2, n),
        'test3': np.random.normal(96, 2, n)}
df = pd.DataFrame(data=data)
```

```python
df = df[(df['test1'] <= 100)]  
df = df[(df['test2'] <= 100)]  
df = df[(df['test3'] <= 100)]
```

```python
def get_label(test3, test2, test1):
    if test1+test2+test3 < 100:
        return 0
    else:
        return 1

df['label'] = df.apply(lambda row: get_label(row['test3'], 
                                             row['test2'], 
                                             row['test1']), axis=1)
```
